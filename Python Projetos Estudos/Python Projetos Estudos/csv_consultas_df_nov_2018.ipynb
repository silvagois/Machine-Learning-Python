```python
# importando as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt

#informar ao matplotlib sobre os graficos inline
%matplotlib inline
```

```python
#Importando algumas colunas do data set
#sep:        separando o arquivo por ponto e virgula
#encoding : informa o padrão utf-8
#usecols   : seleciona as colunas que desejo que apresente no dataframe

#Carregando o arquivo csv
df=pd.read_csv('data set\dados_df_consultas_csv.csv', sep=';',encoding='cp1252',usecols=[0,1,2,8,9,10,12,13])

#Exibindo as primeiras linhas do data frame
df.head(3)
```

```python
#Alterando os nome das colunas
df.columns=['ano_mes','cod_estabelecimento','estabelecimento','cd_procedimento','procedimento','qt',
           'caracter_atendimento','cd_form_organizacao']
df.head(5)
```

```python
#Descrevendo as colunas numericas do dataframe
df.describe()
```

```python
#Listando valores unicos por coluna, trazendo todos estabelecimentos unicamente do data set
df['estabelecimento'].unique()
```

```python
#Quantidade de Consultas feitas por estabelecimento
df['estabelecimento'].value_counts()
```

```python
#Plotar um grafico por procedimento
df['cd_procedimento'].value_counts().plot.bar()
```

```python
#Criando um novo dataframe usando a coluna estabelecimento e filtrando somente pelo hospital materno infantil HMIB
df_estabelecimento = df[df['estabelecimento']=='HOSPITAL MATERNO INFANTIL DE BRASILIA - HMIB']

#linhas iniciais
df_estabelecimento.head()
```

```python
#Gerando um grafico por caracter de atendmento somente do hospital materno infantil de brasilia HMIB
df_estabelecimento['caracter_atendimento'].value_counts().plot.bar()
```

```python
#TRAZENDO SOMENTE AS 5 ULTIMAS LINHAS DO DATA FRAME 
df_estabelecimento.tail()
```

```python
df_estabelecimento['ano_mes'].value_counts().plot.bar()
```

```python
#LISTAR LINHAS ALETATORIA DO DF
df_estabelecimento.sample(5)
```

```python
#listar quantidade de procedimentos no data frame do hospital materno infantil
# O maior procedimento execultado para o HMI é o de exame Exame Citopatológico Cervico Vaginal/microflora-rastreamento
df_estabelecimento['procedimento'].value_counts()
```

```python
#Selecioar no data frame somente os procedimentos igual a dosagem porém selecionando por string
df_estabelecimento[df_estabelecimento['procedimento'].str.contains('Dosagem')].count()
```

```python
#Criando um novo data frame somente para os procedimentos Dosagem de Fosforo 
df_dosagem_fosforo = df[df['procedimento']=='Dosagem de Fosforo']

#Mostrando as primeiras linhas
df_dosagem_fosforo.head()
```

```python
#mostrando as ultimas linhas
df_dosagem_fosforo.tail()
```

```python
#Quais hospitais ou estabeleicimentos mais tem procedimentos de dosagem de fosforo
df_dosagem_fosforo['estabelecimento'].value_counts()
```

```python
# Qual nivel de atendimento mais tem esse atendimento
df_dosagem_fosforo['caracter_atendimento'].value_counts()
```

```python
#Grafico de nivel de urgencia para dosagem de fosforo
#Plotando o grafico na horizonal barh
df_dosagem_fosforo['caracter_atendimento'].value_counts().plot.barh()
```

```python
#Melhorando o grafico de barras invertendo a ordem 
# e colocar um titulo para o grafico
df_dosagem_fosforo['caracter_atendimento'].value_counts(ascending=True).plot.barh(title='Caracter Atendimento')
```

```python
df_dosagem_fosforo['caracter_atendimento'].value_counts().plot.hist(bins=30,edgecolor = 'black',title='asdf')
```
