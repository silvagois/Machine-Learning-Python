```python
# Video desta aula : https://www.youtube.com/watch?v=DNegkxwQJuY&t=545s
#Carregando os dados
#Conjuntos de dados do site kanggle link:https://www.kaggle.com/dell4010/wine-dataset
#Apresentou erro de unicode tiver que alterar as barras \ para /
import pandas as pd
dataset = pd.read_csv('C:/Users/Marcos/Documents/Machine-Learning-Python/Python Projetos Estudos/Python Projetos Estudos/ML Python/Data Set/wine_dataset.csv',sep=',',encoding='cp1252')
```

```python
#Mostra as 5 primeiras linhas
#A coluna Style que informa se o vinho Ã© tinto ou branco 
dataset.head()
```

```python
#Traz as 5 ultimas linhas 
dataset.tail()
```

```python
# Alterando o campo Style red para  0
dataset['style'] = dataset['style'].replace('red',0)
```

```python
# Alterando o campo Style white para 1
dataset['style'] = dataset['style'].replace('white',1)
```

```python
#Confirmando se houve a alteraÃ§Ã£o com sucesso da coluna Style red para zero e white para 1
dataset.head()
```

```python
#MOstra a quantidade de linhas 6497 e 13 colunas do dataset
dataset.shape
```

```python
# Separando as variÃ¡veis enre prditoras e variavel alvo 
# No caso a variÃ¡vel alvo Ã© a coluna style para descobrir se o vinho Ã© tinto ou branco
y = dataset['style'] #alvo
x = dataset.drop('style',axis = 1) #dropando a coluna style, todo o dataset esta sendo armazenado em x exceto a col style
```

```python
#Imprimindo a variÃ¡vel X
x
```

```python
#Imprimindo a variÃ¡vel y
print(y)
```

```python
# Preparando para o treino e teste
#funÃ§Ã£o de treino train_test_split dentro do pacote model_selection
from sklearn.model_selection import train_test_split

#criando um conjunto de dados de treino e teste
x_treino, x_teste, y_treino, y_teste = train_test_split(x,y, test_size = 0.3) #Passando 30% para teste e 70% Treino
```

```python
x_treino.shape
```

```python
y_treino.shape
```

```python
x_teste.shape
```

```python
y_teste.shape
```

```python
# arvove de decisÃ£o algoritmo de classificaÃ§Ã£o 
from sklearn.ensemble import ExtraTreesClassifier
#CriaÃ§Ã£o do MOdelo de treino
modelo = ExtraTreesClassifier()
modelo.fit(x_treino,y_treino)#fit ira aplicar os dados 

#Imprimindo o REsultado ClassificaÃ§Ã£o 
resultado = modelo.score(x_teste, y_teste)
print('Acuracia:', resultado)
```
